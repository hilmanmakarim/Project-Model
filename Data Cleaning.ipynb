```python
listItem = []
for col in df.columns :
  listItem.append([col, df[col].dtype, df[col].isna().sum(), round((df[col].isna().sum() / len(df[col])) * 100,2),
                   df[col].nunique(), list(df[col].unique()[:2])]);

dfDesc = pd.DataFrame(columns = ['dataFeatures', 'dataType', 'null', 'nullPct', 'unique', 'uniqueSample'], data = listItem)
dfDesc
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['contact'].value_counts()
```

```python
# so for "unknown" in Job, Contact and Education, i will replace it into "Other"
df[['job', 'contact', 'education']] = df[['job', 'contact', 'education']].replace(['unknown'],'other')
```

```python
listItem = []
for col in df.columns :
  listItem.append([col, df[col].dtype, df[col].isna().sum(), round((df[col].isna().sum() / len(df[col])) * 100,2),
                   df[col].nunique(), list(df[col].unique()[:2])]);

dfDesc = pd.DataFrame(columns = ['dataFeatures', 'dataType', 'null', 'nullPct', 'unique', 'uniqueSample'], data = listItem)
dfDesc
```
